```python
from collections import Counter
import pandas as pd
from ar_wordcloud import ArabicWordCloud
import nltk
```

```python
df = pd.read_csv("/content/arabic_dataset_classifiction.txt", encoding='utf-8')
print(df)
```

```python
df.drop("targe", axis=1, inplace=True)
```

```python
print(df)
```

```python
awc = ArabicWordCloud(background_color="Black")
```

```python
def word_count(fname):
        with open(fname) as f:
                return Counter(f.read().split())
File=word_count("/content/arabic_dataset_classifiction.txt")
print("Number of words in the file :",word_count("/content/arabic_dataset_classifiction.txt"))
```

```python
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words("arabic")
stopwords
```

```python
dict_wc = awc.from_dict(File, ignore_stopwords=False)
awc.plot(dict_wc, title="تغريدات", width=12, height=12)
```

```python
from ar_wordcloud import ArabicWordCloud
from ar_wordcloud.utils import read_mask_image

mask_img = read_mask_image() # NOTE: pass `mask_img_url` param to use a different mask, the default is a heart
awc = ArabicWordCloud(background_color="Black", mask=mask_img, contour_width=4, scale=0.5)

awc.from_file('/content/arabic_dataset_classifiction.txt').to_image()
```
